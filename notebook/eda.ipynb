```python
num_features = [
    'amt',
    'city_pop',
    'lat', 'long',
    'merch_lat', 'merch_long',
    'distance_km',          
    'txn_time_gap'          
]

low_card_cat = [
    'gender',
    'state',
    'category'
]

high_card_cat = [
    'merchant',
    'cc_num'
]
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("low_cat", OneHotEncoder(handle_unknown="ignore"), low_card_cat),
        ("high_cat", OrdinalEncoder(), high_card_cat)
    ]
)
```

```python
import pandas as pd
import numpy as np

df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'])
```

```python
df['txn_hour'] = df['trans_date_trans_time'].dt.hour
df['txn_dayofweek'] = df['trans_date_trans_time'].dt.dayofweek
df['is_weekend'] = df['txn_dayofweek'].isin([5, 6]).astype(int)
```

```python
df['customer_age'] = (
    (df['trans_date_trans_time'] - df['dob'])
    .dt.days // 365
)
```

```python
from math import radians, cos, sin, asin, sqrt

def haversine(lat1, lon1, lat2, lon2):
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
   
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    return 6371 * c  # km

df['distance_km'] = df.apply(
    lambda x: haversine(
        x['lat'], x['long'],
        x['merch_lat'], x['merch_long']
    ),
    axis=1
)
```

```python
df = df.sort_values(['cc_num', 'trans_date_trans_time'])

df['txn_time_gap'] = (
    df.groupby('cc_num')['trans_date_trans_time']
    .diff()
    .dt.seconds
)

df['txn_time_gap'].fillna(0, inplace=True)
```

```python
# Final fix: Use .rolling('1h').count() on a column that always exists after groupby and set_index
# Here, use 'amt' (transaction amount) which is always present

# If 'trans_date_trans_time' is not a column, reset index to bring it back
if 'trans_date_trans_time' not in df.columns:
    df = df.reset_index()

# Make sure 'trans_date_trans_time' is datetime type
if not np.issubdtype(df['trans_date_trans_time'].dtype, np.datetime64):
    df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])

# Set index to 'trans_date_trans_time'
df = df.set_index('trans_date_trans_time')

# Calculate rolling count per card for 1 hour window using .count() on 'amt' column
# This avoids KeyError because 'amt' is always present

df['txn_count_1h'] = (
    df.groupby('cc_num')
      .rolling('1h')['amt']
      .count()
      .reset_index(level=0, drop=True)
)

# Reset index to default
df = df.reset_index()
```

```python
df['avg_amt_per_card'] = df.groupby('cc_num')['amt'].transform('mean')

df['amt_deviation'] = df['amt'] / (df['avg_amt_per_card'] + 1)
```

```python
df['is_new_merchant'] = (
    df.groupby('cc_num')['merchant']
    .transform(lambda x: ~x.duplicated())
).astype(int)
```

```python
drop_cols = [
    'trans_date_trans_time',
    'unix_time',
    'dob'
  
]

df.drop(columns=drop_cols, inplace=True)
```

```python
X = df.drop(columns=['is_fraud'])
y = df['is_fraud']
```

```python
num_features = [
    'amt',
    'city_pop',
    'lat', 'long',
    'merch_lat', 'merch_long',
    'distance_km',
    'txn_time_gap',
    'txn_count_1h',
    'avg_amt_per_card',
    'amt_deviation',
    'customer_age',
    'txn_hour',
    'is_weekend'
]
low_card_cat = [
    'gender',
    'state',
    'category'
]
high_card_cat = [
    'merchant',
    'cc_num'
]
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('low_cat', OneHotEncoder(handle_unknown='ignore'), low_card_cat),
        ('high_cat', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), high_card_cat)
    ],
    remainder='drop'
)
```

```python
X_processed = preprocessor.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep = preprocessor.transform(X_test)
```

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(
    n_estimators=200,
    contamination=0.01,   # approx fraud rate
    random_state=42,
    n_jobs=-1
)

iso_forest.fit(X_train_prep)
```

```python
iso_preds = iso_forest.predict(X_test_prep)

# Convert to 0/1 fraud labels
iso_preds = [1 if x == -1 else 0 for x in iso_preds]
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, iso_preds))
print(classification_report(y_test, iso_preds))
```

```python
from xgboost import XGBClassifier

# handle heavy class imbalance
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_pos_weight,
    eval_metric="logloss",
    random_state=42,
    n_jobs=-1
)

xgb_model.fit(X_train_prep, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

xgb_preds = xgb_model.predict(X_test_prep)

print(confusion_matrix(y_test, xgb_preds))
print(classification_report(y_test, xgb_preds))
```

```python
from sklearn.metrics import roc_auc_score

xgb_probs = xgb_model.predict_proba(X_test_prep)[:, 1]
roc_auc = roc_auc_score(y_test, xgb_probs)

print("ROC-AUC:", roc_auc)
```
